```python
import numpy as np
```

```python
'''
Newton- Ranpson method of logistic regression:

 W+=(X.T@W@X)^-1@X.T@(y-p)

 where W is the diagonal matrix of weight and p is the predicted probabilities
'''
class LogisticRegression:
    def __init__(self,max_iter):
        self.max_iter=max_iter
        self.weights=None
        self.p=None
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    def predict(self,X):
        weighted_sum=X@self.weights
        prediction=self.sigmoid(weighted_sum)
        return prediction
    def MSE(self,y,p):
        mse=0
        for r,s in zip(y,p):
            mse+=(r-s)**2
        return mse
    def fit(self,X,y):
        X=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
        self.weights=np.zeros(X.shape[1])
        for _ in range(self.max_iter):
            self.p=self.predict(X)
            W=np.diag(self.p*(1-self.p))
            self.weights+=np.linalg.inv(X.T@W@X)@X.T@(y-self.p)
            print(f"Iternation number :{_+1} | MSE : {self.MSE(self.p,y)}\n")
                
```

```python

X_test = np.array([
    [1.0, 1.5],
    [1.2, 2.5],
    [1.5, 1.0],
    [2.0, 2.0],
    [8.0, 8.5],
    [9.0, 9.0],
    [8.5, 9.5],
    [9.5, 8.0]
])

y_test = np.array([0, 0, 0, 0, 1, 1, 1, 1])


model = LogisticRegression(max_iter=10)
model.fit(X_test, y_test)

print("Final weights:", model.weights)
```
